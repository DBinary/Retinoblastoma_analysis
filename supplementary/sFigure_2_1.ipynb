```python
# 绘图备用函数

def plot_text_set(text):
    if len(text)>40:
        ty=text.split(' ')
        ty_len=len(ty)
        ty_mid=ty_len//2
        res=''
        for i in range(ty_len):
            if i!=ty_mid:
                res+=ty[i]+' '
            else:
                res+='\n'+ty[i]+' '
        return res
    else:
        return text

from colormap import Colormap
c = Colormap()
gocmap = c.cmap_linear('#F0F3BD',
 '#FEE582',
 '#F6BD60',)
c.test_colormap(gocmap)
```

## GSEA1

```python
gseada1=data1
#倍数变化规则
gseada1['fcsign']=np.sign(gseada1['log2FC'])
gseada1['metric']=gseada1['-log10pvalue']/gseada1['fcsign']
gseada1.dropna(inplace=True)
gseada1
```

```python
data1.sort_values(['log2FC'],ascending=False)
```

```python
gseada1 = gseada1[~gseada1.index.str.contains('RPS')]
gseada1 = gseada1[~gseada1.index.str.contains('RPL')]
gseada1
```

```python
gseada1_sig = gseada1[gseada1.pvalue < 0.05]
gseada1_up = gseada1_sig[gseada1_sig.log2FC > 0.05]
gseada1_dw = gseada1_sig[gseada1_sig.log2FC < 0]
```

```python
gseada1_up
```

```python
rnk_pos1=pd.DataFrame()
rnk_pos1['gene_name']=gseada1_up.index
rnk_pos1['rnk']=gseada1_up.loc[:,'log2FC'].values
rnk_pos1.dropna()
rnk_pos1.sort_values(by=['rnk'],ascending=False,inplace=True)
rnk_pos1
```

```python
import gseapy as gp
pre_res_pos1 = gp.prerank(rnk=rnk_pos1, # or rnk = rnk,
                     gene_sets='GO_Biological_Process_2021',
                     min_size=5,
                     max_size=1000,
                     permutation_num=1000, # reduce number to speed up testing
                     outdir=None, # don't write to disk
                     seed=6,
                     verbose=True, # see what's going on behind the scenes
                    )
pre_res_pos1.res2d['Term'] = pre_res_pos1.res2d.index
pre_res_pos1.res2d['-log10pvalue'] = -np.log10(pre_res_pos1.res2d['pval']+0.001)
pre_res_pos1
```

```python
pre_res_pos1.res2d['fraction']=pre_res_pos1.res2d['matched_size']/pre_res_pos1.res2d['geneset_size']
pre_res_pos1.res2d
```

```python
plot_data21 = pre_res_pos1.res2d.sort_values('fdr')[:9].sort_values('nes')
plot_data21 = plot_data21.sort_values('fraction')
plot_data21
```

```python
# color
gocmap = c.cmap_linear('#DFEAF6',
        '#9DCBE3',
        '#4292C5',)

pp=plt.figure(figsize=(2,4))
#用ax控制图片
ax=pp.add_subplot(1,1,1)

#圆圈大小控制
a = 10

#plot_data2 = pre_res_pos.res2d.sort_values('fdr')[:9].sort_values('nes')
b=plt.scatter(plot_data21['fraction'],range(len(plot_data21['nes'])),
           s=plot_data21['matched_size']*a,linewidths=1,edgecolors='black',c=plot_data21['nes'],cmap=gocmap)
ax.yaxis.tick_right()
plt.yticks(range(len(plot_data21['fraction'])),[plot_text_set(i.split('(')[0]) for i in plot_data21['Term']],
          fontsize=12,)
plt.xticks(fontsize=12,)
plt.title('DEGs of KEGG Enrichments',fontsize=12)
plt.xlabel('Fractions of genes in KEGG',fontsize=12)


fig = plt.gcf()
cax = fig.add_axes([3, 0.55, 0.6, 0.02])
#im = cax.imshow(plot_data2['logp'], cmap=gocmap)

#绘制图例栏
cb=plt.colorbar(shrink=0.25,cax=cax,orientation='horizontal')
cb.set_label(r'$Normalized$ $Enrichment$ $Score$',fontdict={'size':12})

gll = plt.scatter([],[], s=5*a, marker='o', color='white',edgecolors='black')
gl = plt.scatter([],[], s=7*a, marker='o', color='white',edgecolors='black')
ga = plt.scatter([],[], s=9*a, marker='o', color='white',edgecolors='black')

plt.legend((gll,gl,ga),
       ('5', '7', '9'),
       loc='lower left',
       ncol=3,bbox_to_anchor=(-0.45, -13),
       fontsize=12)
plt.text(0,-20,'Count') #这里counts指捕获这个信号通路的基因数目
plt.savefig("MSCMCKEGG.png",dpi=300,bbox_inches ='tight')
plt.savefig("/content/drive/MyDrive/LIU/Data/MSCMCKEGG.png",dpi=300,bbox_inches ='tight')
```

```python
ndata11.obsm['mlm_estimate']
```

```python
ndata11.obsm['progeny_mlm_estimate'] = ndata11.obsm['mlm_estimate'].copy()
ndata11.obsm['progeny_mlm_pvals'] = ndata11.obsm['mlm_pvals'].copy()
ndata11
```

```python
acts = dc.get_acts(ndata11, obsm_key='mlm_estimate')
acts
```

```python
mean_acts = dc.summarize_acts(acts, groupby= 'Type', min_std=0)
mean_acts
```
